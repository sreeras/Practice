# Generate all possible sorted arrays from alternate elements of two given sorted arrays
#Given two sorted arrays A and B, generate all possible arrays such that first element is taken from A then #from B then from A and so on in increasing order till the arrays exhausted. The generated arrays should end #with an element from B.

For Example 
A = {10, 15, 25}
B = {1, 5, 20, 30}

The resulting arrays are:
  10 20
  10 20 25 30
  10 30
  15 20
  15 20 25 30
  15 30
  25 30


A = {10, 15, 25, 28}
B = {1, 5, 20, 30}

The resulting arrays are:
  10 20
  10 20 25 30
  10 20 28 30
  10 30
  15 20
  15 20 25 30
  15 30
  25 30

```python
def genAllPos(A,B):

    finalList = []
    for i in range(len(A)):
        finalList += genAllPosRec(A, B, i, 0) #########10,2 15, 2, 25,3?
        
    print (finalList)

def genAllPosRec(A, B, startinA, startinB, retList = [[]]):

    if (startinB == len(B)):
        return retList

    if A[startinA] < B[startinB]:
        lastB = B[startinB]
        retList.append(retList[-1] + [A[startinA], B[startinB]])
        
        while startinA < len(A) and A[startinA] < lastB:
            startinA += 1
        
        if startinA >= len(A):
            return retList
    
    while A[startinA] > B[startinB]:
        startinB += 1
          
    genAllPosRec(A, B, startinA, startinB, retList)
    
    return retList
```
