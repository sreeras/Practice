```python
with open("hello.txt", 'w+') as fh:
    
    fh.writelines('\n'.join(map(str, range(1,11))))

with open("hello.txt",'r') as fh:
    
    for line in fh.readlines():
        print (line, end='')

with open("hellobinary.txt","wb+") as fh:
    
    for i in range(10):
        if i % 2:
            fh.write(bytes('Harsha','utf-8'))
        else:
            fh.write(bytes('Sam','utf-8'))
        
```

```python
def print_all_files(dPath=''):
    
    for file in os.listdir():
        if not dPath:
            dPath = os.getcwd()
        full_path = os.path.join(dPath, file)
        
        if os.path.isdir(full_path):
            print_all_files(full_path)
        else:
            print(full_path)
```

```python
print_all_files()
```
